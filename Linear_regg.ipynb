# Y= Mx+b
# price = M * Area + B
### here price is dependable variable and area is an independable variable
### m = Slope B= Intercept

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df =pd.read_csv(r"C:\Users\srika\OneDrive\Documents\OneDrive\Desktop\dj\linear_regression.csv")
df
```

```python
print(df.columns)
```

```python
plt.xlabel('area(sqft)')
plt.ylabel('PRICE (usd)')
plt.scatter(df.Area,df.Price,color="red")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Area']],df.Price)
```

```python
reg.predict([[5200],[3600],[2600]]) 
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.xlabel('area(sqft)', fontsize=20)
plt.ylabel('PRICE (usd)',fontsize=20)
plt.scatter(df.Area,df.Price,color="red")
plt.plot(df.Area,reg.predict(df[['Area']]),color='blue')
```
