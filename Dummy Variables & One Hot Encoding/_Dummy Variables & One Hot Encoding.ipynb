```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\srika\OneDrive\Documents\OneDrive\Desktop\dj\Dummy Variables & One Hot Encoding\carprices.csv")
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
dummies = dummies.astype(int)
dummies
```

```python
mearge = pd.concat([df,dummies], axis = 'columns')
mearge
```

```python
final = mearge.drop(['Car Model','Mercedez Benz C class'],axis='columns')
final
```

```python
x= final.drop(['Sell Price($)'],axis='columns')
x
```

```python
y = final['Sell Price($)']
y
```

```python
model=linear_model.LinearRegression()
model.fit(x,y)
```

### predict price of a mercedez benz that is 4 year old with mileage 45000

```python
model.predict([[4500,4,0,0]])
```

### predict price of a BMW X5 that  is 7year Old With milleage 86000

```python
model.predict([[86000,7,0,1]])
```

### the score (accuracy)

```python
model.score(x,y)
```
